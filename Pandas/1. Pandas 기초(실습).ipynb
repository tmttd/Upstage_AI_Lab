## 2.3. 여러 DataFrame 합치기

```python
## 아래 표는 'A'라는 col에 있는 값들이 'A0', 'A1', 'A2', 'A3'이라는 뜻

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A1', 'A2'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C6', 'C7', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# 위-아래로 합치기(순서대로 합치기)
pd.concat([df1, df2, df3])

# axis =1 즉 행 기준으로 합치면 index를 기준으로 합치기 때문에 NaN이 이렇게 발생함.
# df1과 df2와 df3의 index는 서로 다 다름.
## 따라서 행(row, 가로)으로 합치고 싶으면 index를 전부 통일시켜야 함.

pd.concat([df1, df2, df3], axis=1)
```

```python
# index reset 코드 (index가 전부 0부터 시작하는 것으로 초기화 됨.)
## drop=True를 넣으면 기존 index가 떨어져 버림.

df2.reset_index(drop=True) # view. 보여주기만 하는 거임. 편집된 거 X

### 그런데 이렇게 원본 데이터를 수정하는 코드를 사용해도 원본은 수정되지 않음.
#### 일시적으로 수정해서 보여주기만 함. 원본에 진짜 수정하고 싶으면 덮어쓰기 해야 함.
```

```python
pd.concat([df1, df2.reset_index(drop=True), df3.reset_index(drop=True)], axis=1)

## 깔끔하게 붙이기 성공!
```

```python
# SQL과 같이 join operation을 사용할 수 있다.

## join operation은 예를 들어, 학생 A1 A2의 정보를 합치고 싶을 때 사용함.

# A(left) inner join B(right) on 'A'
# --> table A와 table B의 column A에서 겹치는 값들을 기준으로 두 테이블을 합치는 연산.

pd.merge(df2,
         df3,
         on='C',
         how='inner')
```

```python
# left join: left(df1)을 기준으로 right table(df2)에서 on에 대해서 겹치는 대상을 붙임.
# 겹치지 않으면 NaN으로 붙여 줌.
pd.merge(df1, df2, on="A", how='left')
```

```python
# train.csv 파일 불러오기
titanic = pd.read_csv(r'C:\Users\user\Desktop\Upstage AI Lab\EDA lecture\Python EDA\Pandas\data\train.csv')
titanic
```

```python
# Q1. 여성 승객들의 평균 나이를 계산해보세요.
titanic.Sex == 'female'

titanic.loc[titanic.Sex == 'female', 'Age']

# Q2. 1등석에 탑승한 승객 중 최대 요금을 낸 사람의 이름을 찾아주세요.

titanic.Pclass == 1

temp = titanic.loc[titanic.Pclass == 1, ['Fare', 'Name']]
temp.loc[temp.Fare == temp.Fare.max(), ['Fare', 'Name']]

# Q3. 1등석에 탑승한 승객 중 승선한 항구 도시가 Queenstown인 사람들의 수는?

temp2 = titanic.loc[titanic.Pclass == 1, 'Embarked']
temp2.loc[titanic.Embarked == 'Q'].count()

## mask & mask : 두 조건을 만족하는 mask.
## mask | mask : 두 조건 중 하나라도 만족하는 mask.
### 개수 세는 것은 len을 쓰거나, 끝에 shape[0]을 붙이기! count()는 모든 칼럼
#### 을 카운트해서 쓸 데 없이 길어짐.

titanic[(titanic.Pclass == 1) & (titanic.Embarked == 'Q')]

# Q4. 승선한 곳이 S인 사람들의 생존자 수는?
## 아래 것은 ==을 사용하는 것과 똑같다. 살짝 연산이 다르긴 함. 확장 가능.
### 조건을 여러 개 달면서 확장 가능. 

titanic.Embarked.isin(['S', 'Q'])

titanic.loc[titanic.Embarked.isin(['S']), 'Survived'].sum()
#### sum()을 해버리면, 생존자는 1로 표시되기 때문에 1이 전부 더해져서
#### 생존자 수가 나와버림. Pandas에서 자주 쓰는 기술임.

# Q5. 미혼 여성 중 나이를 모르는 사람의 수는?

## pandas의 str 안에 문자열 함수가 많이 있음.

titanic.loc[titanic.Name.str.contains('Miss.'), 'Age'].isnull().sum()
```

## 2.5. Pivot Table을 이용하여 데이터 살펴보기

- pivot table이란 기존 테이블 구조를 특정 column을 기준으로 재구조화한 테이블을 말합니다.

- 특정 column을 기준으로 pivot하기 때문에, 어떤 column에 어떤 연산을 하느냐에 따라서 만들어지는 결과가 바뀝니다.

- 주로 어떤 column을 기준으로 데이터를 해석하고 싶을 때 사용합니다.

```python
# 성별을 기준으로 생존률 파악 --> Mean vs Sum
pd.pivot_table(data=titanic, index='Sex', values='Survived', aggfunc=['count', 'sum', 'mean'])
```

```python
# 사회 계급을 기준으로 생존률 파악
## index, aggfunc는 전부 여러 개 설정할 수 있음.

pd.pivot_table(data=titanic, index=['Sex', 'Pclass'], values='Survived', aggfunc=['count', 'sum', 'mean']).plot()
```
