## 1. Titanic data load 및 환경 세팅 

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot') # ggplot style을 사용합니다.
```

```python
# 타이타닉 데이터 불러오기
train = pd.read_csv("data/train.csv")
train
```

- titanic 생존 여부 분석에 필요하지 않은 column들은 없애줍니다.

```python
# 어떤 column들이 분석에 필요하지 않을까요? # PassengerId, Name, Ticket
#train = train.drop(columns=["PassengerId", "Name", "Ticket"])
train.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True)
train
```

- 연산이 불가능한 범주형 정보들(categorical feature)을 변환해줍니다.

```python
train.info() # DataFrame에서 object라는 Dtype을 가지는 Feature들은 대부분 categorical feature입니다.
```

```python
# 범주형 column을 찾아보고, one-hot encoding을 이용해 변경합니다.

# 1) One-hot encoding : 하나의 정보만 hot(1)로 만드는 인코딩 방식.

#             --->  Male Female Not Responding
# Female             0     1          0
# Male               1     0          0
# Not Responding     0     0          1

# 2) Label encoding (v)
# 3) Ordinal encoding
            
# Female         --->      0
# Male                     1
# Not Responding           2
df = pd.get_dummies(train, columns=["Sex", "Embarked"]) # One-hot encoding 적용
df
```

## 2. Pivot table을 이용한 feature별 데이터 분석 

```python
pd.pivot_table(df, index="Fare")
```

```python
# 성별 정보를 위주로 데이터를 바라보기 위해 pivot table 생성
pd.pivot_table(train, index=["Sex"]) # mean
```

```python
grouped = pd.pivot_table(df, index=["Sex_male", "Sex_female"])
#pd.pivot_table(df, index=["Embarked_C", "Embarked_Q", "Embarked_S"])
```

```python
# bar plot
grouped.plot(kind="barh")
```

### 2.1. Pivot Table with multi-index (동시에 여러 feature의 관점에서 보기)

```python
# 성별과 승객등급별에 따라 pivot table을 생성합니다.
pd.pivot_table(train, index=["Sex", "Pclass"])
pd.pivot_table(df, index=["Sex_male", "Pclass"])
```

### 2.2. Aggregation function in Pivot Tables

- 얻은 pivot table에 원하는 통계량을 계산합니다. e.g. mean(), sum(), min() 등등

```python
# 성별/승객등급을 기준으로 만든 pivot table에서 age에 대해서 평균값을, survived에 대해서 sum값을 계산합니다.
pd.pivot_table(train, index=["Sex", "Pclass"], aggfunc={"Age":np.mean, "Survived":np.sum})
```

- values= 옵션을 통해 특정 column의 value를 지정할 수 있습니다.

```python
grouped = pd.pivot_table(train, index=["Sex", "Pclass"], values=["Fare"], aggfunc=np.sum)
grouped
```

```python
# bar plot
grouped.plot(kind="barh")
```

- columns= 를 통해서 생존여부에 대한 정보를 위주로 pivot table이 형성될 수 있게 지정한다.

```python
grouped = pd.pivot_table(train, index=["Sex"], columns=["Survived"], values=["Age"])
grouped
```

```python
# bar plot
grouped.plot(kind="bar")
```

### 2.3. Handling missing data in Pivot Tables

- 실제 데이터셋을 다룰때, 비어있는 정보(없어졌거나, 얻지 못한)에 대한 처리를 꼭 해주어야합니다. 정보가 비어있으면 전처리 이후에 ML 방법들을 사용할 수 없기 때문입니다.

```python
pd.pivot_table(train, index=["Sex", "Pclass"], values=["Survived"])
```

```python
grouped = pd.pivot_table(train, index=["Sex", "Survived", "Pclass"],
                        columns=["Embarked"], values=["Age"],
                         fill_value=train["Age"].mean())
grouped
```

- pandas pivot_table에서는 2가지 방법으로 NaN값을 처리할 수 있다.


1) dropna - drops all null values in pivot table


2) fill_value - replace all null values in pivot table with the specified values

- 또는 날려버리지 않고, 평균이나 최소/최대값을 사용하는 경우도 있다.
