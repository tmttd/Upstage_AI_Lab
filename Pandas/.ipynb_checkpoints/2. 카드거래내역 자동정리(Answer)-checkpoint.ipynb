```python
import pandas as pd
```

```python
data = pd.read_excel("data/2020_02_거래내역.xls")#.dropna()
data
```

```python
stores = list(data["가맹점명"])
price = list(data["이용금액"])
categories = list(data["카테고리"])
```

```python
print(len(data["가맹점명"])) # 61이어야 합니다.
print(len(set(data["가맹점명"])))
```

```python
target = pd.read_excel("data/2020_03_거래내역.xls").dropna().reset_index(drop=True)
target
```

```python
target_categories = [] # 3월 거래내역에 추가될 "카테고리" 정보.
target_stores = list(target["가맹점명"])
target_price = list(target["이용금액"])

for store in target_stores:
    """
    3월 거래내역에 있는 가맹점명을 순서대로 하나씩 보면서,
    해당 가맹점명이 포함된 2월달 거래내역의 카테고리를 3월 거래내역에 추가하면 됩니다.
    
    3월 첫번째 거래내역 -> 2월 거래내역의 몇번째(index)인지 확인 -> 해당 index의 category를 가져옴.
    -> target_categories에 추가. -> .....
    """
    #print(store)
    # "지에스25 뉴서강대학사점"
    try:
        idx = stores.index(store) # store라는 원소가 stores라는 리스트에 어디(index)에 있는지 알려줍니다.
        temp = categories[idx] # 2월달 해당 가맹점명을 가지는 거래내역의 카테고리
    except ValueError:
        temp = "미정" # 3월 거래내역의 가맹점명이 2월 거래내역에 존재하지 않는 경우.
        
    target_categories.append(temp)
```

```python
target["카테고리"] = target_categories
```

```python
target.to_excel("data/거래내역(정리).xlsx")
```
